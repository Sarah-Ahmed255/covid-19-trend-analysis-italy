```python
import pandas as pd
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Total rows and columns
print("Shape of Dataset", df.shape)

# Columns names
print("Columns in Dataset", df.columns)
```

```python
italy_df = df[df['location'] == 'Italy']

# Shows the data of italy only
italy_df.head()
```

```python
# Take only the useful columns
italy_df = italy_df[['date', 'new_cases', 'total_cases', 'new_deaths', 'total_deaths']]

# Convert text date into proper date format
italy_df['date'] = pd.to_datetime(italy_df['date'])

italy_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Adjust figure size
plt.figure(figsize=(12,6))

# Plot line graph for new cases(seaborn)
sns.lineplot(data=italy_df, x='date', y='new_cases', color='blue')

plt.title('Italy - Daily new COVID-19 cases', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('New Cases', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,6))

# Create line plot
sns.lineplot(data=italy_df, x='date', y='new_deaths', color='red')

# Give title and axis names
plt.title('Italy - Daily new Covid-19 deaths', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.xlabel('New Deaths', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(data=italy_df, x='date', y='total_cases', color='green')

plt.title('Italy - Total Covid-19 cases over time ', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.xlabel('Total Cases', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(data=italy_df, x='date', y='total_deaths', color='black')

plt.title('Italy - Total Covid-19 deaths over time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.xlabel('Total Deaths', fontsize=12)
plt.show()
```

```python
# New cases and deaths comparison
plt.figure(figsize=(14, 6))

sns.lineplot(data=italy_df, x='date', y='new_cases', label='New Cases', color='blue')

sns.lineplot(data=italy_df, x='date', y='new_deaths', label='New Deaths', color='red')

plt.title('Italy - New COVID-19 Cases & Deaths over time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend()
plt.show()
```

```python
# Total cases and deaths comparison
plt.figure(figsize=(14, 6))

sns.lineplot(data=italy_df, x='date', y='total_cases', label='Total Cases', color='green')
sns.lineplot(data=italy_df, x='date', y='total_deaths', label='Total Deaths', color='black')

plt.title('Italy - Total COVID-19 Cases & Deaths over time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend()
plt.show()
```
